# Data Cleaning and EDA

## Removing irrelavant columns

```python
#Removing certain columns that are not relevant

df.drop(['description','url','image_url', 'region_url'],axis=1,inplace= True)
df.shape
```

## Removing irrelevant rows

### retaining only properties with 1 to 4 bedrooms and bathrooms

```python
#removing rows with more than 4 bedrooms or zero bedrooms
df1 = df[(df.beds==1)|(df.beds==2)|(df.beds==3)|(df.beds==4)]
df1.shape
```

```python
#removing rows with more than 4 bathrooms or zero bathrooms
df2 = df1[(df1.baths==0.5)|(df1.baths==1)|(df1.baths==1.5)|(df1.baths==2)|(df1.baths==2.5)|(df1.baths==3)|(df1.baths==3.5)|(df1.baths==4)]
df2.shape
```

### Finding and removing outliers

```python
#finding outliers using IQR method

Q1= df2['price'].quantile(0.25)
Q3 = df2['price'].quantile(0.75)   

IQR = Q3 - Q1

upper_limit = Q3 + 1.5 * IQR
print("upper limit is",upper_limit)
lower_limit = Q1 - 1.5 * IQR
print("lower limit is",lower_limit)    

#finding values less than lower limit and values greater than upper limit
outliers = df2[(df2['price'] < lower_limit) | (df2['price'] > upper_limit)]
print(outliers)
```

```python
#creating a new dataframe without the outliers
df3 = df2[(df2['price'] > lower_limit) & (df2['price'] < upper_limit)]

print(df3['price'].max())
print(df3['price'].min())
```

```python
#finding outliers using IQR method for sqfeet column

Q1= df3['sqfeet'].quantile(0.25)
Q3 = df3['sqfeet'].quantile(0.75)   

IQR = Q3 - Q1

upper_limit = Q3 + 1.5 * IQR
print("upper limit is",upper_limit)
lower_limit = Q1 - 1.5 * IQR
print("lower limit is",lower_limit)    

#finding values less than lower limit and values greater than upper limit
outliers = df3[(df3['sqfeet'] < lower_limit) | (df3['sqfeet'] > upper_limit)]
print(outliers)
```

```python
#creating a new dataframe without the outliers
df4 = df3[(df3['sqfeet'] > lower_limit) & (df3['sqfeet'] < upper_limit)]

print(df4['sqfeet'].max())
print(df4['sqfeet'].min())
```

# Machine learning 

## Preparing dataframe for machine learning algorithms

```python
#Removing certain columns that are not relevant
cleaned_df4 = cleaned_df.copy()
cleaned_df4.drop(['id','area','state','price','sqfeet','lat','long'],axis=1,inplace= True)
cleaned_df4.shape
```

```python
cleaned_df4.columns
```

## Applying One Hot Encoding for categorical variables

```python
#Make copy to avoid changing original data 
encoded_data = cleaned_df4.copy()

from sklearn.preprocessing import OneHotEncoder

categorical_variables = ['type','region','laundry_options', 'parking_options']


#Using one-hot encoding to convert the categorical variables into numerical
encoded_data = pd.get_dummies(data=encoded_data, columns=categorical_variables)
encoded_data
```

```python
encoded_data.columns
```

```python
#change order of columns to have the dependent variable as last column
ML_df = encoded_data.reindex(columns = ['beds', 'baths','type_apartment',
       'type_condo', 'type_duplex', 'type_house', 'type_manufactured',
       'type_other', 'type_townhouse', 'region_midwest', 'region_north_east',
       'region_south', 'region_west','comes_furnished',
       'sqrt_sqfeet', 'wheelchair_access', 'electric_vehicle_charge',  'laundry_options_laundry',
       'laundry_options_no laundry on site', 'laundry_options_not specified',
       'laundry_options_washer/dryer', 'parking_options_carport',
       'parking_options_garage', 'parking_options_no parking',
       'parking_options_not specified', 'parking_options_off-street parking',
       'parking_options_street parking','cats_allowed', 'dogs_allowed', 'smoking_allowed','sqrt_price'])

ML_df
```

## SelectKBest Feature Selection

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression

#selecting the independent variables
Xx = ML_df.iloc[:,0:-1]  
#selecting the target column
yy = ML_df.iloc[:,-1]     
yy = yy.astype('int')

#apply SelectKBest class to extract top 25 best features
bestfeatures = SelectKBest(score_func=f_regression, k=10)
fit = bestfeatures.fit(Xx,yy)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(Xx.columns)

#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  
print(featureScores.nlargest(26,'Score'))  
```

```python
#removing the columns that were not included in the best features
ML_df.drop(['type_apartment','type_duplex','parking_options_street parking', 'dogs_allowed'], axis=1, inplace=True) 
ML_df.shape
```

```python
ML_df.head()
```

```python
ML_df.columns
```

## Splitting the dataset into train and test 

```python
from sklearn.model_selection import train_test_split

array = ML_df.values

# Features: columns without the dependent variable
X = array[:,0:-1]

# Target variable: seperating the dependent variable 
y = array[:,-1]

# 20% of the data will be used for testing
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.2, random_state= 88)
```

## Applying machine learning algorithms

### Decision tree regressor

```python
from sklearn.tree import DecisionTreeRegressor 
from sklearn import metrics

#fitting model 
DT =  DecisionTreeRegressor()
DT.fit(X_train, y_train)

#predicting values 
DT_predictions = DT.predict(X_test)

print('Decision Tree regression Performance:')

print('MAE:', metrics.mean_absolute_error(y_test, DT_predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, DT_predictions)))
print('R2_Score: ', metrics.r2_score(y_test, DT_predictions))

#plot the graph of feature importances 
feat_importances_DT = pd.Series(DT.feature_importances_, index=ML_df.columns[:-1])
feat_importances_DT.nlargest(10).plot(kind='barh')
plt.title("Decision Tree Regressor Feature Importance")
plt.show()
```

```python
sns.scatterplot(x = DT_predictions, y = y_test);
```

```python
#comparing the actual and predicted values in a plot
y_test = list(y_test)
plt.plot(y_test, color = 'blue', label = 'Actual Value')
plt.plot(DT_predictions, color = 'orange', label = 'Predicted Value')
plt.title('Actual Price Vs Predicted Price')
plt.xlabel('No. of values')
plt.ylabel('Rental Price')
plt.legend()
plt.show()
```

## Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

#defining hyperparameters 
RandomForestRegressor_params = {'n_estimators': 500,'min_samples_split': 6, 'min_samples_leaf': 1,'bootstrap': True}

#fitting RF model
RF = RandomForestRegressor(**RandomForestRegressor_params)
RF.fit(X_train, y_train)

#predicting values 
RF_predictions = RF.predict(X_test)

print('Random forest Performance :')

print('MAE:', metrics.mean_absolute_error(y_test, RF_predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, RF_predictions)))
print('R2_Score: ', metrics.r2_score(y_test, RF_predictions))

#plot the graph of feature importances 
feat_importances_RF = pd.Series(RF.feature_importances_, index=ML_df.columns[:-1])
feat_importances_RF.nlargest(10).plot(kind='barh')
plt.title("Random Forest Regressor Feature Importance")
plt.show()
```

```python
sns.scatterplot(x = RF_predictions, y = y_test);
```

```python
#comparing the actual and predicted values in a plot
y_test = list(y_test)
plt.plot(y_test, color = 'blue', label = 'Actual Value')
plt.plot(RF_predictions, color = 'orange', label = 'Predicted Value')
plt.title('Actual Price Vs Predicted Price')
plt.xlabel('No. of values')
plt.ylabel('Rental Price')
plt.legend()
plt.show()
```

## Extreme Gradient Boosting Regressor

```python
from xgboost import XGBRegressor

#defining hyperparameters 
XGBRegressor_params = {'n_estimators': 4500,'max_depth': 8,'learning_rate': 0.1,'random_state': 101}

#creating an instance of gradient boosting regressor
XGBRegressor = XGBRegressor(**XGBRegressor_params)

#fitting GBR model before scaling
XGBRegressor.fit(X_train, y_train)

#predicting values before scaling
XGBRegressor_predictions = XGBRegressor.predict(X_test)

print('Gradient Boosting Regressor Performance:')

print('MAE:', metrics.mean_absolute_error(y_test, XGBRegressor_predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, XGBRegressor_predictions)))
print('R2_Score: ', metrics.r2_score(y_test, XGBRegressor_predictions))

#plot the graph of feature importances 
feat_importances_XGB = pd.Series(XGBRegressor.feature_importances_, index=ML_df.columns[:-1])
feat_importances_XGB.nlargest(10).plot(kind='barh')
plt.title("XGB Regressor Feature Importance")
plt.show()
```

## K-nearest Neighbors Regressor 

```python
from sklearn import neighbors
from sklearn.metrics import mean_squared_error
from math import sqrt
%matplotlib inline

#empty list to store each rmse value
rmse_val = [] 
for K in range(10):
    K = K+1
    model = neighbors.KNeighborsRegressor(n_neighbors = K)
    
    model.fit(X_train, y_train)  
    pred=model.predict(X_test) 
    error = sqrt(mean_squared_error(y_test,pred)) 
    rmse_val.append(error) 
    print('RMSE value for k= ' , K , 'is:', error)
```

```python
from sklearn.model_selection import GridSearchCV
params = {'n_neighbors':[2,3,4,5,6,7,8]}

knn = neighbors.KNeighborsRegressor()

model = GridSearchCV(knn, params, cv=5)
model.fit(X_train,y_train)
model.best_params_ 
```

```python
from sklearn.neighbors import KNeighborsRegressor

#fitting model 
KNN = KNeighborsRegressor(n_neighbors=5)
KNN.fit(X_train, y_train)

#predicting values 
KNN_predictions = KNN.predict(X_test)

print('KNN Performance:')

print('MAE:', metrics.mean_absolute_error(y_test, KNN_predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, KNN_predictions)))
print('R2_Score: ', metrics.r2_score(y_test, KNN_predictions))
```
